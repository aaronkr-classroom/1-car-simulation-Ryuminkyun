```python
# 라이브러리 불러오기
import numpy as np
import matplotlib.pyplot as plt
import math
```

```python
time = np.linspace(
    start = 0,
    stop= 2*math.pi,
    num=100,
    dtype= np.float32
)
y= np.sin(time)
```

```python
plt.figure()
plt.plot(time, y)
plt.xlabel("Time")
plt.ylabel("Signal")
plt.show()
```

```python
x, y, scale = np.random.randn(3, 100)

fig, ax = plt.subplots()
ax.scatter(x = x, y=y,c=scale,
           s=np.abs(scale)*500
           )
ax.set(title = "random plot")
plt.show()
```
